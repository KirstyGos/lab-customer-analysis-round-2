### 7. BONUS: Put all the previously mentioned data transformations into a function 

```python
def transform_data(df):
    df.columns = [df.columns[i].lower().replace(' ', '_') for i in range(len(df.columns))]
    df = df.rename(columns={"employmentstatus": "employment_status"})
    df = df.rename(columns={"unnamed:_0": "customer_number"})
    df = df.dropna(axis=0)
    df["month"] = pd.DatetimeIndex(df["effective_to_date"]).month
    df_1st_quarter = df.where(df['month'] < 4)
    return df_1st_quarter

df = pd.read_csv("files_for_lab/csv_files/marketing_customer_analysis.csv")
transform_data(df)
```
